```python
# Install necessary libraries
!pip install librosa tensorflow scikit-learn pandas

import pandas as pd
import os
import numpy as np
import librosa
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical
from google.colab import drive
import warnings
warnings.filterwarnings('ignore')

# --- IMPORTANT: VERIFY THESE PATHS ---
# Mount your Google Drive
drive.mount('/content/drive', force_remount=True)

# Path to your main 'sound' dataset folder in Google Drive
# Based on your screenshots, this path looks correct.
SOUND_DATASET_ROOT = "/content/drive/MyDrive/19AI702/sound/sound/"

# Define the categories and their corresponding numerical labels
# We'll treat 'ambulance' as 1, and 'traffic' (and 'firetruck') as 0 (not a siren)
CLASS_LABELS = {
    "ambulance": 1,
    "traffic": 0,
    "firetruck": 0 # Assuming firetruck sirens are also to be distinguished from ambulance
}

# Feature extraction parameters
N_MFCC = 40 # Number of MFCC features to extract
```

```python
def load_audio_data(root_path, subfolder_type):
    """
    Loads audio file paths and assigns labels from a given root path and subfolder type (e.g., 'train' or 'test').
    """
    filepaths = []
    labels = []

    # Path to either the 'train' or 'test' directory within the main sound dataset
    data_dir = os.path.join(root_path, subfolder_type)

    # Iterate through each class folder (ambulance, firetruck, traffic)
    for class_name, label in CLASS_LABELS.items():
        class_path = os.path.join(data_dir, class_name)

        if not os.path.isdir(class_path):
            print(f"Warning: Directory not found: {class_path}. Skipping this class for {subfolder_type}.")
            continue

        for filename in os.listdir(class_path):
            if filename.endswith(('.wav', '.mp3', '.m4a', '.flac')): # Add other audio extensions if needed
                filepaths.append(os.path.join(class_path, filename))
                labels.append(label)

    return filepaths, labels

print("--- Loading Training Data ---")
train_filepaths, train_labels = load_audio_data(SOUND_DATASET_ROOT, 'train')
print(f"Loaded {len(train_filepaths)} training audio files.")

print("\n--- Loading Testing Data ---")
test_filepaths, test_labels = load_audio_data(SOUND_DATASET_ROOT, 'test')
print(f"Loaded {len(test_filepaths)} testing audio files.")

# No balancing needed here as you provide distinct folders for train/test already
```

```python
def extract_features(file_path, n_mfcc=N_MFCC):
    """Extracts MFCCs from an audio file."""
    try:
        audio, sample_rate = librosa.load(file_path, res_type='kaiser_fast')
        mfccs = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=n_mfcc)
        mfccs_scaled = np.mean(mfccs.T, axis=0) # Take the mean of MFCCs over time
        return mfccs_scaled
    except Exception as e:
        print(f"Error processing {os.path.basename(file_path)}: {e}")
        return None

# Extract features for training data
X_train_features = []
y_train_encoded = []

print("\n--- Extracting Features for Training Data ---")
for i, filepath in enumerate(train_filepaths):
    features = extract_features(filepath)
    if features is not None:
        X_train_features.append(features)
        y_train_encoded.append(train_labels[i])

# Extract features for testing data
X_test_features = []
y_test_encoded = []

print("\n--- Extracting Features for Testing Data ---")
for i, filepath in enumerate(test_filepaths):
    features = extract_features(filepath)
    if features is not None:
        X_test_features.append(features)
        y_test_encoded.append(test_labels[i])

# Convert lists to numpy arrays
X_train = np.array(X_train_features)
X_test = np.array(X_test_features)

# Convert labels to one-hot encoding (e.g., 0 -> [1,0], 1 -> [0,1])
y_train = to_categorical(y_train_encoded, num_classes=len(CLASS_LABELS))
y_test = to_categorical(y_test_encoded, num_classes=len(CLASS_LABELS))

print(f"\nTraining features shape: {X_train.shape}, Training labels shape: {y_train.shape}")
print(f"Testing features shape: {X_test.shape}, Testing labels shape: {y_test.shape}")
```

```python
# Build the model
model = Sequential([
    Dense(256, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.5),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(len(CLASS_LABELS), activation='softmax') # Output layer matches number of classes
])

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()

# Train the model
print("\n--- Starting Model Training ---")
history = model.fit(X_train, y_train,
                    epochs=100, # You can adjust epochs
                    batch_size=32,
                    validation_data=(X_test, y_test),
                    verbose=2)

print("\n--- Training Finished ---")
```

```python
# Save the trained model to a file
model.save('siren_classifier.h5')

print("Model saved as siren_classifier.h5")
```
